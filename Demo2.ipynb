```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
base_url = "https://rwandahouseland.com/forSale"
response = requests.get(base_url)
response.status_code
```

```python
soup = BeautifulSoup(response.text, "html.parser")  # ✅ No comma
# soup= (response.text,"lxml")
soup
```

```python

combination = soup.find_all('div', class_="relative group bg-white rounded-lg overflow-hidden shadow-sm hover:shadow-md transition-all duration-300")

name = soup.find_all('h5', class_='text-sm font-medium text-gray-900 truncate')
price = soup.find_all('div', class_='flex items-center gap-2')  
location =soup.find_all('p', class_='text-gray-600') 
category = soup.find_all('span', class_='text-xs text-gray-600 truncate')

# print("Name:", name)
# print("Price:", price)
# print("Location:", location)
# print("Category:", category)

combination
```

```python
houses = soup.find_all('div', class_='relative group bg-white rounded-lg overflow-hidden shadow-sm hover:shadow-md transition-all duration-300')


data = []

for house in houses:
    name = house.find('h5', class_='text-sm font-medium text-gray-900 truncate')
    price = house.find('div', class_='flex items-center gap-2')
    category = house.find('span', class_='text-xs text-gray-600 truncate')
    location = house.find('span', class_='flex items-center gap-2')

    name = name.get_text(strip=True) if name else 'N/A'
    price = price.get_text(strip=True) if price else 'N/A'
    location = location.get_text(strip=True) if location else 'N/A'
    category = category.get_text(strip=True) if category else 'N/A'

    data.append({
        'Name': name,
        'Price': price,
        'Location': location,
        'Category': category
    })



df = pd.DataFrame(data)

print(df)
```

```python
# import requests
# from bs4 import BeautifulSoup
# import pandas as pd

all_data = []

for page in range(1, 5):
    url = 'https://rwandahouseland.com/forSale?page={page}'
    print("Getting page {page}...")

    response = requests.get(url)  
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        houses = soup.find_all('div', class_='relative group bg-white rounded-lg overflow-hidden shadow-sm hover:shadow-md transition-all duration-300')

        for house in houses:
            name = house.find('h5', class_='text-sm font-medium text-gray-900 truncate')
            price = house.find('div', class_='flex items-center gap-2')
            category = house.find('span', class_='text-xs text-gray-600 truncate')
            location = house.find('span', class_='flex items-center gap-2')

            all_data.append({
                'Name': name.get_text(strip=True) if name else 'N/A',
                'Price': price.get_text(strip=True) if price else 'N/A',
                'Location': location.get_text(strip=True) if location else 'N/A',
                'Category': category.get_text(strip=True) if category else 'N/A'
            })
    else:
        print("Skipping page {page} due to error: status code {response.status_code}")

# Save to CSV
df = pd.DataFrame(all_data)
df.to_csv('rwandahouses.csv', index=False)

print("Done! Data saved to rwandahouses.csv")
```
